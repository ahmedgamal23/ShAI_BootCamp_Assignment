```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
df.columns
```

# 1 - Basic Data Exploration:


```python
# Identify the number of rows and columns in the dataset, determine the data types of each column, and check for missing values in each column.
print("**********************************************")
print("Identify the number of rows and columns in the dataset : ", df.shape)
print("**********************************************")
print("determine the data types of each column : ")
print(df.info())
print("**********************************************")
print("check for missing values in each column. : ")
print(df.isna().sum())
print("**********************************************")
```

# 2 - Descriptive Statistics:

```python
# Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.
print("**********************************************")
print("mean, median, mode, minimum, and maximum salary : ")
print(np.mean(df["TotalPay"]))
print(np.median(df["TotalPay"]))
print(df["TotalPay"].mode())
print(np.min(df["TotalPay"]))
print(np.max(df["TotalPay"]))
print("**********************************************")
print("determine the range of salaries : ")
print(np.max(df["TotalPay"]) - np.min(df["TotalPay"]))
print("**********************************************")
print("find the standard deviation. : ")
print(np.std(df["TotalPay"]))
print("**********************************************")
```

# 3 - Data Cleaning:

```python
# Handle missing data by suitable method with explain why you use it.
# 1 - the column Notes and Status is null so drop this columns from data
df = df.drop(["Notes", "Status"], axis=1)

# 2 - and other nan columns fill it by mean
df = df.fillna({'BasePay':df['BasePay'].mean(), 'OvertimePay':df['OvertimePay'].mean(), 'OtherPay':df['OtherPay'].mean(),'Benefits':df['Benefits'].mean()} )
```

```python
df.isna().sum()
```

# 4 - Basic Data Visualization:

```python
import matplotlib.pyplot as plt
# Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.
# Plotting a histogram
plt.figure(figsize=(8, 6))
plt.hist(df['TotalPay'], bins=10, color='skyblue', edgecolor='black')
plt.title('TotalPay Distribution Histogram')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# use pie charts to represent the proportion of employees in different departments.
plt.figure(figsize=(8,6))
plt.pie(df["JobTitle"].value_counts(), labels=df["JobTitle"].value_counts().index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# 5 - Grouped Analysis:

```python
# Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.
print("********************************************")
print("Group the data by one or more columns and calculate summary statistics for each group : ")
df_group = df.groupby("JobTitle")
avg_per_group = df_group['BasePay'].mean()
print(avg_per_group)
print("********************************************")
print("compare the average salaries across different groups. : ")
avg_per_group.sort_values()
```

# 6 - Simple Correlation Analysis:

```python
# Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.
cor = df.corr()
print(cor)
plt.scatter(x=df["TotalPay"], y=df["BasePay"])
plt.xlabel("Salary")
plt.ylabel("BasePay")
plt.show()
```
